```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
form= LabelEncoder()
df['Form'] = form.fit_transform(df['Form'])
form
```

```python
categorical_columns = ['Fruit_x','Fruit_y','Form', 'CupEquivalentUnit_x','CupEquivalentUnit_y', 'RetailPriceUnit_x','RetailPriceUnit_y']
label_encoders = {}
for col in categorical_columns:  # assume this is a list of column names
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
```

```python
X=df.drop(['RetailPrice_x'], axis=1)
y=(df[['RetailPrice_x']])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=101)
```

```python
import xgboost as xgb

model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)
```
